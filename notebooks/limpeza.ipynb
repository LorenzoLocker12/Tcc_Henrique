## Notebook para Limpeza do Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

```

```python
# Importa o Dataset Utilizando a engine 'calamine' para importação mais rápida
df = pd.read_excel('../data/raw/matriz_alunos.xlsx', engine='calamine')
```

```python
# Exibir o Dataset
df.head()
```

```python
# Seleciona todas as colunas, exceto 'Aluno'
cols_to_transform = df.columns.drop('Aluno')

# Lista de valores que devem ser convertidos para 0
valores_para_zerar = ['', ' ']

# Substitui 'X' por 1 e 'M' por 0 nas colunas selecionadas
df[cols_to_transform] = df[cols_to_transform].replace({'X': 1, 'M': 0})

# Seleciona as colunas (exceto 'Aluno') e substitui os valores da lista por 0
df[cols_to_transform] = df[cols_to_transform].replace(valores_para_zerar, 0)

# Como uma garantia final, preenche quaisquer valores nulos (NaN) verdadeiros com 0
df[cols_to_transform] = df[cols_to_transform].fillna(0)

# Substitui qualquer valor nulo (NaN) restante por 0
df = df.fillna(0)

# Transformar tudo para Int
df[cols_to_transform] = df[cols_to_transform].astype(int)
```

```python
# Exportar para a pasta correta
df.to_excel('../data/processed/matriz_alunos.xlsx', index=False)
```
